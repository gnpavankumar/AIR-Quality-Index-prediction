```python
import pandas as pd
data=pd.read_csv('/content/city_day.csv')
data
```

```python
data.isnull().sum()
```

```python
data['PM2.5'].fillna(data['PM2.5'].mean(),inplace=True)
data['PM10'].fillna(data['PM10'].mean(),inplace=True)
data['SO2'].fillna(data['SO2'].mean(),inplace=True)
data['NO2'].fillna(data['NO2'].mean(),inplace=True)
data['NO'].fillna(data['NO'].mean(),inplace=True)
data['CO'].fillna(data['CO'].mean(),inplace=True)
data['O3'].fillna(data['O3'].mean(),inplace=True)
data['AQI'].fillna(data['AQI'].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python

newdata=data.drop(['City', 'Date','NOx', 'NH3','Benzene', 'Toluene', 'Xylene', 'AQI_Bucket'],axis=1)
newdata.columns
```

```python
newdata.isnull().sum()
```

```python
newdata.info()
```

```python
prepareddata=newdata.astype(int)
prepareddata.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
x=prepareddata['AQI']
y1=prepareddata['PM2.5']
y2=prepareddata['PM10']
y3=prepareddata['NO']
y4=prepareddata['NO2']
y5=prepareddata['CO']
y6=prepareddata['SO2']
y7=prepareddata['O3']
plt.figure(figsize=(15,8))
plt.scatter(x,y1,label='PM 2.5',color='salmon')
plt.scatter(x,y2,label='PM 10',color='palegreen')
plt.scatter(x,y3,label='NO',color='yellow')
plt.scatter(x,y4,label='NO2',color='steelblue')
plt.scatter(x,y5,label='CO',color='lime')
plt.scatter(x,y6,label='SO2',color='violet')
plt.scatter(x,y7,label='O3',color='springgreen')
plt.title('AQI and its Pollutents',fontsize=18)
plt.xlabel('AQI',fontsize=14)
plt.ylabel('Value',fontsize=14)
plt.legend()
plt.show()
corr = prepareddata.corr()
sns.heatmap(corr, annot=True)
```

```python
from sklearn.model_selection import train_test_split

x=prepareddata[['PM2.5', 'PM10', 'NO', 'NO2','CO', 'SO2','O3']]
y=prepareddata['AQI']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
```

```python
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
print("Accuracy of the Decision Tree Regressor  model comes to be: \n ")
print(model.score(x_train,y_train))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
mae = mean_absolute_error(y_test, pred)
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, pred)


print("Regression Report for Decision Tree Regressor")
print("---------------------------------------------")
print(f"Mean Absolute Error (MAE):      {mae:.4f}")
print(f"Mean Squared Error (MSE):       {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R² Score:                        {r2:.4f}")
```

```python
import joblib


joblib.dump(model, 'aqi_decision_tree_model.pkl')
print("Model saved successfully!")

```
